```python
import pandas as pd
```

```python
# ===== 1. Đọc file =====
input_path = r"D:\User\Tài liệu học\HK2 - 2025-2026\4. Sentiment score\Sentiment_Score_LLM.csv"   # đổi lại đường dẫn nếu cần
df = pd.read_csv(input_path)
```

```python
# ===== 2. Đảm bảo cột điểm là số =====
df["senti_score"] = pd.to_numeric(df["senti_score"], errors="coerce").fillna(0)
```

```python
# ===== 3. Xoay chiều: mỗi Review_ID = 1 dòng, mỗi code = 1 cột =====
df_wide = (
    df.pivot_table(
        index="review_id",        # gom các clause cùng review vào 1 dòng
        columns="code",           # mỗi code thành 1 cột
        values="senti_score",     # lấy điểm sentiment
        aggfunc="sum",            # cộng các điểm của cùng code trong cùng review
        fill_value=0              # code nào không có thì điền 0
    )
    .reset_index()
)
```

```python
# ===== 4. Nếu muốn sắp xếp cột theo thứ tự cố định =====
desired_order = ["review_id", "UH", "CT", "TCr", "LC", "VE", "SE", "RC", "EC", "RA", "VR", "TC", "RP", "EL", "HL", "CA"]
existing_order = [col for col in desired_order if col in df_wide.columns]
other_cols = [col for col in df_wide.columns if col not in existing_order]
df_wide = df_wide[existing_order + other_cols]
```

```python
# ===== 6. Tạo đường dẫn file output cùng thư mục với file input =====
output_path = "Matrix_wide.csv"
df_wide.to_csv(output_path, index=False, encoding="utf-8-sig")
print("Đã lưu file:", output_path)
```
