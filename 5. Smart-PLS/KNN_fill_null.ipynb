```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.impute import KNNImputer
```

```python
# ===== 1. Đọc file =====
input_path = Path(r"D:\User\Tài liệu học\HK2 - 2025-2026\Smart-PLS\Matrix_wide.csv")   # nếu cần thì thay bằng full path
df = pd.read_csv(input_path)
```

```python
# ===== 2. Tách cột ID và cột điểm =====
id_col = "Review_ID"
score_cols = [col for col in df.columns if col != id_col]
```

```python
# ===== 3. Đổi 0 thành NaN để KNN hiểu đó là giá trị thiếu =====
# CHỈ dùng dòng này nếu 0 trong file của bạn là "thiếu", không phải điểm thật
df_knn = df.copy()
df_knn[score_cols] = df_knn[score_cols].replace(0, np.nan)
```

```python
# ===== 4. KNN Imputation với k = n =====
imputer = KNNImputer(n_neighbors=14)
df_knn[score_cols] = imputer.fit_transform(df_knn[score_cols])
```

```python
# ===== 5. Ghép lại với Review_ID =====
df_filled = df_knn.copy()
```

```python
# ===== 6. Lưu file ra cùng thư mục với file gốc =====
output_path = input_path.parent / f"{input_path.stem}_knn14_filled.csv"
df_filled.to_csv(output_path, index=False, encoding="utf-8-sig")
```

```python
# ===== 7. Kiểm tra =====
print("Đã lưu file tại:", output_path.resolve())
print("\nSố giá trị thiếu trước khi fill:")
print(df[score_cols].replace(0, np.nan).isna().sum())

print("\nSố giá trị thiếu sau khi fill:")
print(df_filled[score_cols].isna().sum())

print("\n5 dòng đầu sau khi fill:")
print(df_filled.head())
```
