```python
import pandas as pd
import glob
import os

# 1. Thêm dấu \ vào sau chữ 'Dữ liệu' và dùng 'r' để tránh lỗi
path = r'D:\User\File python\Dữ liệu nhóm\*.csv' 

files = glob.glob(path)

# Kiểm tra xem có thấy file không
print(f"Tìm thấy {len(files)} file CSV.")

if len(files) == 0:
    print("Vẫn không thấy file. Hãy thử kiểm tra lại tên thư mục 'Dữ liệu' có dấu cách ở cuối không nhé.")
else:
    li = []
    for filename in files:
        # Đọc file - thêm encoding để tránh lỗi font tiếng Việt
        df = pd.read_csv(filename, index_col=None, header=0, encoding='utf-8-sig')
        li.append(df)

    # Gộp dữ liệu
    frame = pd.concat(li, axis=0, ignore_index=True)

    # Làm sạch dữ liệu khoảng trắng ở cột address (nếu có) trước khi lọc
    if 'address' in frame.columns:
        frame['Address'] = frame['Address'].astype(str).str.strip()

    # 2. Sắp xếp theo địa điểm (A-Z) và sao (Giảm dần 5 -> 1)
    # Thay 'address' và 'stars' bằng tên cột thật trong file của bạn
    frame = frame.sort_values(by=['Address', 'Rating'], ascending=[True, False])

    # 3. Lọc trùng
    frame = frame.drop_duplicates()

    # 5. Lưu kết quả
    frame.to_csv(r'D:\User\File python\raw_data.csv', index=False, encoding='utf-8-sig')
    print("Đã gộp và lọc xong! File 'ket_qua_tong_hop.csv' nằm trong thư mục 'File python'.")
```
