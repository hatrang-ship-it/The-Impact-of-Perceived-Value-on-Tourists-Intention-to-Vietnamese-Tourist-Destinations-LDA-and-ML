```python
import pandas as pd
```

```python
# 1. Đọc dữ liệu từ file Excel (thay 'data.xlsx' bằng tên file của bạn)
df = pd.read_csv(r'D:\User\File python\Data_nhom.csv')
```

```python

# 1 & 2. Nhóm dữ liệu theo 'Address' và 'Rating', sau đó đếm số lượng
thong_ke = df.groupby(['Address', 'Rating']).size().reset_index(name='Số lượng Comment')

# Tạo một list để chứa các phần dữ liệu sẽ ghép lại
danh_sach_df = []

# Duyệt qua từng địa điểm
for dia_diem, group in thong_ke.groupby('Address'):
    # Đưa các dòng chi tiết của địa điểm này vào list
    danh_sach_df.append(group)
    
    # Tính tổng số comment của địa điểm đó
    tong_dia_diem = group['Số lượng Comment'].sum()
    
    # Tạo một dòng tổng cộng (chỉ 1 dòng)
    dong_tong = pd.DataFrame({
        'Address': [dia_diem], 
        'Rating': ['Tổng cộng'], # Thay Rating bằng chữ 'Tổng cộng' cho dễ nhìn
        'Số lượng Comment': [tong_dia_diem]
    })
    
    # Đưa dòng tổng vào list (nằm ngay dưới các dòng chi tiết)
    danh_sach_df.append(dong_tong)

# Nối tất cả các DataFrame trong list lại thành một bảng hoàn chỉnh
ket_qua_cuoi_cung = pd.concat(danh_sach_df, ignore_index=True)

# 3. Xuất kết quả
print("Thống kê số lượng comment theo địa điểm và rating:")
# Dùng .to_string() để in ra TẤT CẢ các dòng, không bị ẩn bởi dấu "..."
print(ket_qua_cuoi_cung.to_string())

# Tổng tất cả dữ liệu
tong_tat_ca = thong_ke['Số lượng Comment'].sum()
print(f"\nTổng số dữ liệu của cả file là: {tong_tat_ca}")
```

```python
# Tính tổng tất cả dữ liệu
tong_tat_ca = thong_ke['Số lượng Comment'].sum()

# Tạo một DataFrame chứa dòng tổng cộng cuối cùng
dong_tong_tat_ca = pd.DataFrame({
    'Address': ['TỔNG CỘNG TẤT CẢ'], 
    'Rating': [''], # Để trống cột Rating cho dòng này
    'Số lượng Comment': [tong_tat_ca]
})

# Nối dòng tổng cộng này vào cuối dataframe hiện tại
ket_qua_cuoi_cung = pd.concat([ket_qua_cuoi_cung, dong_tong_tat_ca], ignore_index=True)
ket_qua_cuoi_cung.to_csv(r'D:\User\File python\thongke.csv', index=False, encoding='utf-8-sig')
```
