```python
import pandas as pd
import joblib
from pathlib import Path
```

```python
# =========================
# 2. KHAI BÁO ĐƯỜNG DẪN FILE
# =========================
input_path = Path(r"D:\User\Tài liệu học\HK2 - 2025-2026\1. Data\5. Split clauses\processed_data_translated_en_split_clauses.csv")
model_path = Path(r"D:\User\Tài liệu học\HK2 - 2025-2026\3. Label sample set\LDA_topic\Full_labeled\File parameter của mô hình\lr_model.pkl")
vectorizer_path = Path(r"D:\User\Tài liệu học\HK2 - 2025-2026\3. Label sample set\LDA_topic\Full_labeled\File parameter của mô hình\lr_vectorizer.pkl")
```

```python
# File output sẽ lưu cùng thư mục với file input
output_path = input_path.parent / f"{input_path.stem}_labeled.csv"
```

```python
# =========================
# 3. ĐỌC DỮ LIỆU
# =========================
df = pd.read_csv(input_path)

print("Số dòng dữ liệu:", len(df))
print("Các cột hiện có:", df.columns.tolist())
```

```python
# =========================
# 4. CHỌN CỘT CHỨA CLAUSE
# =========================
TEXT_COL = "Clause"

if TEXT_COL not in df.columns:
    raise ValueError(f"Không tìm thấy cột '{TEXT_COL}' trong file dữ liệu.")

# Xử lý text rỗng / NaN
texts = df[TEXT_COL].fillna("").astype(str)
```

```python
# =========================
# 5. LOAD MODEL VÀ VECTORIZER
# =========================
model = joblib.load(model_path)
vectorizer = joblib.load(vectorizer_path)

print("Đã load model và vectorizer thành công.")
```

```python
# =========================
# 6. VECTORIZE CLAUSES
# =========================
# Quan trọng: dùng transform, KHÔNG dùng fit_transform
X = vectorizer.transform(texts)

print("Shape sau khi vectorize:", X.shape)
```

```python
# =========================
# 7. DỰ ĐOÁN TOPIC / CODE
# =========================
df["predicted_topic"] = model.predict(X)

print("Đã dự đoán topic xong.")
```

```python
# =========================
# 8. LƯU FILE KẾT QUẢ
# =========================
df.to_csv(output_path, index=False, encoding="utf-8-sig")

print("Đã gán topic xong.")
print("File đã lưu tại:", output_path.resolve())

# Xem thử kết quả
print(df[[TEXT_COL, "predicted_topic"]].head(10))
```
