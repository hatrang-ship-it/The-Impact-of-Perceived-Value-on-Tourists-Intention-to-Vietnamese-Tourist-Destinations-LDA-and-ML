```python
# =========================
# 1. IMPORT THƯ VIỆN
# =========================
import pandas as pd
from pathlib import Path
```

```python
# =========================
# 2. KHAI BÁO FILE INPUT
# =========================
INPUT_PATH = Path(r"D:\User\Tài liệu học\HK2 - 2025-2026\1. Data\5. Split clauses\processed_data_translated_en_split_clauses.csv")

OUTPUT_PATH = INPUT_PATH.with_name(INPUT_PATH.stem + "_sample_10pct.csv")
```

```python
# =========================
# 3. ĐỌC FILE
# =========================
df = pd.read_csv(INPUT_PATH)

print("Số dòng ban đầu:", len(df))
print("Các cột trong file:")
print(df.columns.tolist())
```

```python
# =========================
# 4. LẤY NGẪU NHIÊN 10% DÒNG DỮ LIỆU
# =========================
RANDOM_STATE = 42

df_sample = df.sample(
    frac=0.10,
    random_state=RANDOM_STATE
).copy()

df_sample = df_sample.reset_index(drop=True)
```

```python
# =========================
# 5. LƯU FILE MỚI
# =========================
df_sample.to_csv(OUTPUT_PATH, index=False, encoding="utf-8-sig")
```

```python
# =========================
# 6. KIỂM TRA KẾT QUẢ
# =========================
print("Số dòng sample 10%:", len(df_sample))
print("File sample đã lưu tại:", OUTPUT_PATH.resolve())

display(df_sample.head(20))
```
